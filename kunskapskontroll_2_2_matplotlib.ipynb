```python
import numpy as np
import matplotlib.pyplot as plt

# --- Data för scatter plot ---
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)

# --- Data för bar plot ---
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# --- Skapa figur med två subplots bredvid varandra ---
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))  # 1 rad, 2 kolumner

# --- Första subplot: Scatter plot ---
ax1.scatter(random_data_x, random_data_y, alpha=0.5, label="Random points")
ax1.plot(x, y, color='red', label="y = x²")  # Valfritt, som extra referens
ax1.set_title("Scatter Plot")
ax1.set_xlabel("X-värden")
ax1.set_ylabel("Y-värden")
ax1.legend()
ax1.grid(True)

# --- Andra subplot: Bar plot ---
ax2.bar(names, values, color='skyblue')
ax2.set_title("Fruit Count")
ax2.set_xlabel("Fruit Type")
ax2.set_ylabel("Count")
ax2.grid(axis='y')

# --- Anpassa layout och visa ---
plt.tight_layout()
plt.show()
```
